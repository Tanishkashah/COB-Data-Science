```python

import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTSS-TcErkXNk8KB0AlijhitwetxeHD2M3R0HJl2QPMAyFq0fxFX4PFKnzAWLDnratIz67DNL6GsZnV/pub?output=csv&urp=gmail_link'
df = pd.read_csv(url)
```

```python
# Replace missing values with the mean of the column
df = df.fillna(df.mean())
```

```python
# Remove outliers using the IQR method
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
# Define the range for outliers
outlier_range = 1.5 * IQR
```

```python
# Identify the outliers
outliers = df[(df < (Q1 - outlier_range)) | (df > (Q3 + outlier_range))]
```

```python
# Remove the outliers
df = df[~df.isin(outliers).any(axis=1)]
```

```python
# Save the cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)
```
